```python
# TensorFlow and tf.keras
import tensorflow as tf

# Helper libraries
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import regularizers
from PIL import Image
import numpy as np
import tensorflow as tf
from tensorflow.keras.applications import VGG16
from tensorflow.keras.layers import Flatten, Dense, Dropout
from tensorflow.keras.models import Sequential, load_model
```

```python
# Chemin vers le répertoire contenant les images d'entraînement
train_dir = r"C:\\Users\\marti\\OneDrive\\Martin OneDrive\\ISEP\\A2\\Semestre 2\\Ai and optimisation\\Projet\train\\Test 28-04-2023\\Entrainement\\"

# Charger les noms des fichiers d'images d'entraînement et leurs étiquettes
train_images = []
train_labels = []

class_names = ['Bee', 'Bumblebee', 'Butterfly', 'Dragonfly', 'Hoverfly',
               'Wasp']

for label, class_name in enumerate(os.listdir(train_dir)):
    class_dir = os.path.join(train_dir, class_name)
    for filename in os.listdir(class_dir):
        img_path = os.path.join(class_dir, filename)
        img = Image.open(img_path)
        img = img.resize((256, 256))
        img = img.convert("RGB")
        img_array = np.array(img) / 255.0  # Convertir en tableau NumPy et normaliser les valeurs des pixels
        train_images.append(img_array)
        train_labels.append(label)
        print(img_path)


# Étendre les dimensions des images pour ajouter un canal pour la couleur (RGB)
train_images2 = np.expand_dims(train_images, axis=-1)

train_labels_np = np.array(train_labels)

# Charger le modèle VGG16 pré-entraîné sans les couches fully connected (top) et sans charger les poids
base_model = VGG16(weights="vgg16_weights_tf_dim_ordering_tf_kernels_notop.h5", include_top=False, input_shape=(256, 256, 3))


# Geler les poids des couches convolutionnelles pour empêcher l'entraînement
base_model.trainable = False

# Créer un modèle séquentiel en ajoutant le modèle VGG16 et les couches fully connected au-dessus
modelVGG16 = Sequential([
    base_model,
    Flatten(),
    Dense(1024, activation='relu'),
    Dropout(0.2),
    Dense(512, activation='relu'),
    Dropout(0.2),
    Dense(256, activation='relu'),  # Nouvelle couche Dense
    Dropout(0.2),                     # Nouvelle couche Dropout
    Dense(128, activation='relu'),  # Nouvelle couche Dense
    Dropout(0.2),                     # Nouvelle couche Dropout
    Dense(6, activation='linear')
])

# Compiler le modèle
modelVGG16.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

# Afficher un résumé du modèle
modelVGG16.summary()

# Entraîner le modèle avec les données d'entraînement
modelVGG16.fit(train_images2, train_labels_np, epochs=30)
```

```python
# Chemin vers le répertoire contenant les images de test
test_dir = r"C:\\Users\\marti\\OneDrive\\Martin OneDrive\\ISEP\\A2\\Semestre 2\\Ai and optimisation\\Projet\train\\Test 28-04-2023\\IMG_ZX\\"

# Charger les images de test
test_images = []
test_filenames = []
for filename in os.listdir(test_dir):
    img_path = os.path.join(test_dir, filename)
    img = Image.open(img_path)
    img = img.resize((256, 256))
    img = img.convert("RGB")
    img_array = np.array(img) / 255.0
    test_images.append(img_array)  
    test_filenames.append(filename)
    
test_images = np.array(test_images)
```

```python
probability_modelVGG16 = tf.keras.Sequential([modelVGG16, 
                                         tf.keras.layers.Softmax()])
```

```python
predictions = probability_modelVGG16.predict(test_images)
```

```python
print(predictions)
```

```python
import pandas as pd

# Créer une liste pour stocker les résultats
resultsVGG16 = []

# Prédire chaque classe pour chaque image
for i, (img, filename) in enumerate(zip(test_images, test_filenames)):
    # Ajouter l'image à un lot où elle est le seul membre
    imgVGG16 = np.expand_dims(img, 0)
    
    # Prédire la classe pour cette image
    predictions_singleVGG16 = probability_modelVGG16.predict(imgVGG16)
    print(predictions_singleVGG16)
    
    # Trouver l'index de la classe prédite
    predicted_class_indexVGG16 = np.argmax(predictions_singleVGG16[0])
    
    # Récupérer le pourcentage de chance que cela soit cette classe
    confidenceVGG16 = predictions_singleVGG16[0][predicted_class_indexVGG16] * 100
    print(confidenceVGG16)
    
    # Ajouter les résultats à la liste
    resultsVGG16.append([filename, class_names[predicted_class_indexVGG16], confidenceVGG16])
    
# Créer un DataFrame à partir des résultats
df = pd.DataFrame(resultsVGG16, columns=['Nom du fichier', 'Classe prédite', 'Pourcentage de confiance'])

# Enregistrer le DataFrame dans un fichier Excel
df.to_excel('predictionsVGG16.xlsx', index=False)
```
